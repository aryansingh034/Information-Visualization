```python
#data sampling
import random
population =list(range(1,450))
sample_size=15
random_sample=random.sample(population,sample_size)
print("Random sample:",random_sample)
```

```python
#starfield sampling
import numpy as np
import pandas as pd

data=pd.DataFrame({
    'category': ['a','b','c','d'] * 25,
    'value': np.random.randn(100)
})
startified_sample=data.groupby('category',group_keys=False).apply(lambda x:x.sample(2))
print(startified_sample)
```

```python
#cluster sampling
clusters=[list(range(1,26)),list(range(26,51)),list(range(51,76))]
cluster_sample=random.choice(clusters)
print("Cluster sample:",cluster_sample)
```

```python
#data summarization
import pandas as pd
import numpy as np

data=pd.DataFrame({
    'values':np.random.randn(1578)
})
summary=data.describe()
print(summary)
```

```python
#data visual
import matplotlib.pyplot as plt
data['values'].hist(bins=50,edgecolor='magenta')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Values')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris=load_iris()
pca=PCA(n_components=2)
reduced_data=pca.fit_transform(iris.data)

plt.scatter(reduced_data[:,0],reduced_data[:,1],c=iris.target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.show()
```

```python
#parallel cordinates
from pandas.plotting import parallel_coordinates
data=pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/iris-data.csv")
parallel_coordinates(data,'class',color=['b','g','r'])
plt.title('Parallel Coordinates Plot')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.show()
```

```python
#time series visual

import plotly.express as px
import pandas as pd

data=pd.DataFrame({
    'time':[2000,2001,2002,2003,2004,2005],
    'value':[100,200,150,300,250,400],
    'country':['usa']*6
})

fig=px.line(data,x='time',y='value',color='country',title='Time Series Plot')
fig.show()
```
